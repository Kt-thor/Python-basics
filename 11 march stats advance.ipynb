Q2: Differentiate between one-tailed and two-tailed tests.

One-tailed and two-tailed tests are two types of hypothesis tests that differ in the directionality of the hypothesis being tested.

A one-tailed test is a hypothesis test where the null hypothesis is tested against a directional alternative hypothesis. In other words, a one-tailed test tests whether the observed effect is either significantly greater than or significantly less than the null hypothesis value. A one-tailed test is used when there is prior evidence or a strong theoretical basis to suggest that the effect is in a specific direction.

For example, suppose a researcher wants to test whether a new medication reduces the average weight of patients. A one-tailed hypothesis would be "The new medication reduces the average weight of patients." The one-tailed test would then be designed to determine whether the medication reduces the weight of patients significantly.

A two-tailed test is a hypothesis test where the null hypothesis is tested against a non-directional alternative hypothesis. In other words, a two-tailed test tests whether the observed effect is significantly different from the null hypothesis value in either direction. A two-tailed test is used when there is no prior evidence or theoretical basis to suggest that the effect is in a specific direction.

For example, suppose a researcher wants to test whether a new advertising campaign increases the average sales of a product. A two-tailed hypothesis would be "The new advertising campaign has an effect on the average sales of the product." The two-tailed test would then be designed to determine whether the advertising campaign has a significant effect on sales in either direction.

Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.

A Type 1 error is an error that occurs when the null hypothesis is rejected when it is actually true. In other words, a Type 1 error occurs when the hypothesis test falsely indicates the presence of a significant effect or difference between two groups or variables. This error is also known as a false positive. The probability of making a Type 1 error is denoted by the alpha level or the significance level, which is typically set at 0.05 or 0.01.

An example scenario of a Type 1 error is as follows: A researcher is testing whether a new drug is effective in treating a specific disease. The null hypothesis is that the drug has no effect on the disease. If the researcher rejects the null hypothesis based on the hypothesis test, but in reality, the drug has no effect on the disease, it is a Type 1 error.

A Type 2 error is an error that occurs when the null hypothesis is not rejected when it is actually false. In other words, a Type 2 error occurs when the hypothesis test fails to detect a significant effect or difference between two groups or variables when such an effect or difference actually exists. This error is also known as a false negative. The probability of making a Type 2 error is denoted by the beta level, which is typically set at 0.2 or 0.1.

An example scenario of a Type 2 error is as follows: A researcher is testing whether a new drug is effective in treating a specific disease. The null hypothesis is that the drug has no effect on the disease. If the researcher fails to reject the null hypothesis based on the hypothesis test, but in reality, the drug is effective in treating the disease, it is a Type 2 error.



Q4: Explain Bayes's theorem with an example.

Bayes's theorem is a mathematical formula that describes the relationship between conditional probabilities of events. It provides a way to update the probability of an event based on new information or evidence.

The formula for Bayes's theorem is as follows:

P(A|B) = P(B|A) * P(A) / P(B)

where P(A|B) is the probability of event A given that event B has occurred, P(B|A) is the probability of event B given that event A has occurred, P(A) is the prior probability of event A, and P(B) is the prior probability of event B.

To illustrate Bayes's theorem, consider the following example:

Suppose a doctor is trying to diagnose a patient with a rare disease. The prevalence of the disease in the population is 1%, which means that the prior probability of the patient having the disease is 0.01 or 1%.

The doctor orders a diagnostic test for the disease, which has a sensitivity of 90% (i.e., the probability of the test correctly detecting the disease when it is present is 0.9) and a specificity of 95% (i.e., the probability of the test correctly ruling out the disease when it is not present is 0.95).

If the test comes back positive, what is the probability that the patient has the disease?

Using Bayes's theorem, we can calculate the probability as follows:

P(Disease|Positive Test) = P(Positive Test|Disease) * P(Disease) / P(Positive Test)

where P(Disease|Positive Test) is the probability of the patient having the disease given that the test is positive, P(Positive Test|Disease) is the probability of the test being positive given that the patient has the disease (i.e., the sensitivity of the test), P(Disease) is the prior probability of the patient having the disease, and P(Positive Test) is the probability of the test being positive.

Substituting the values into the formula, we get:

P(Disease|Positive Test) = 0.9 * 0.01 / (0.9 * 0.01 + 0.05 * 0.99)

which simplifies to:

P(Disease|Positive Test) = 0.15 or 15%

Therefore, the probability of the patient having the disease given a positive test result is 15%, which is much lower than the probability of a false positive. This result emphasizes the importance of considering both the sensitivity and specificity of a diagnostic test when interpreting its results.

Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

A confidence interval is a range of values that is likely to contain the true population parameter with a certain level of confidence. It provides a measure of the precision of an estimate and is commonly used in inferential statistics to estimate the population parameter based on a sample.

To calculate a confidence interval, we need to specify a confidence level, which is the probability that the interval contains the true population parameter. The most commonly used confidence levels are 90%, 95%, and 99%.

The formula for a confidence interval is:

CI = x̄ ± z * (s / √n)

where CI is the confidence interval, x̄ is the sample mean, z is the z-score corresponding to the desired confidence level, s is the sample standard deviation, and n is the sample size.

For example, suppose we want to estimate the mean weight of a population of fish. We take a random sample of 100 fish and measure their weights. The sample mean weight is 1.5 pounds, and the sample standard deviation is 0.3 pounds.

We want to calculate a 95% confidence interval for the population mean weight. The z-score corresponding to a 95% confidence level is 1.96 (which can be obtained from a z-table or calculated using software).

Substituting the values into the formula, we get:

CI = 1.5 ± 1.96 * (0.3 / √100)

which simplifies to:

CI = 1.5 ± 0.0588

The 95% confidence interval for the population mean weight is therefore 1.4412 to 1.5588 pounds. This means that if we repeated the sampling process many times, 95% of the resulting confidence intervals would contain the true population mean weight.
